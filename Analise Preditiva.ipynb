```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Dados pre processados Relatório Questionário Socioeconômico 1-2022.csv')
```

```python
feature_transform = {}

for col in df.drop(columns=['Inscrição']).keys():
    feature_transform[col] = df[col].unique()
```

```python
from pycaret.anomaly import *
```

```python
s = setup(data = df.drop(columns=['Inscrição']), session_id=123,verbose = False, html = False, 
             ordinal_features = feature_transform
        )
```

```python
s.X_transformed.head()
```

```python
s.X_transformed.tail()
```

```python
anomaly = create_model(model='cluster', verbose=False)
```

```python
plot_model(
    model=anomaly,
    plot='tsne'
)
```

```python
df_anomaly_detect = assign_model(anomaly)
```

```python
df_anomaly_detect.head()
```

```python
df_anomaly_detect.loc[df_anomaly_detect['Anomaly'] == 1].shape
```

```python
df_anomaly_detect['ID'] = df['Inscrição']
filtro = df_anomaly_detect['Anomaly'] != 1
df_anomaly_detect = df_anomaly_detect.loc[filtro]
```

```python
df_anomaly_detect = df_anomaly_detect.drop(columns=['Anomaly','Anomaly_Score'])
```

```python
df_anomaly_detect.shape
```

```python
feature_transform = {}

for col in df_anomaly_detect.drop(columns='ID').keys():
    feature_transform[col] = df_anomaly_detect[col].unique()
```

```python
from pycaret.clustering import *
```

```python
s = setup(data=df_anomaly_detect, verbose=False, html=False, ordinal_features=feature_transform, normalize=True, normalize_method='minmax')
```

```python
s.X_transformed
```

```python
cluster_model = create_model(
    model='kmeans',
    verbose=False,
    num_clusters=3
)

plot_model(
    model=cluster_model,
    plot='cluster',
    feature='ID'
)
```

```python
df_clusterizado = assign_model(model= cluster_model)
```

```python
ohe = setup(data=df_clusterizado.drop(columns=['ID']), verbose= False, html=False, normalize=False)
```

```python
ohe.X_transformed
```

```python
from sklearn.metrics import DistanceMetric

dist = DistanceMetric.get_metric('hamming')
```

```python
distancias = dist.pairwise(ohe.X_transformed)

centrais = []

for l in distancias:
    centrais.append((np.mean(l), np.median(l)))

print(
    f"""Minimo: {min(centrais)}
Maximo: {max(centrais)}
Média: {np.mean(centrais)}
Mediana: {np.median(centrais)}"""
)
```
